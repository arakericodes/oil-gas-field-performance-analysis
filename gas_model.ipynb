```python
import pandas as pd
```

```python
gas_model_dataset = pd.read_csv("heatmap_data.csv")
```

```python
gas_model_dataset.drop("Unnamed: 0", axis=1, inplace=True)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
gas_model_dataset.columns
```

```python
features = ['Operating cost']
target = ['Gas(MCF)']
```

```python
x = gas_model_dataset[features]
y = gas_model_dataset[target]
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.25, random_state=42)
```

```python
lim = LinearRegression()
lim.fit(x_train, y_train)
y_pred = lim.predict(x_test)
```

```python
from sklearn.metrics import r2_score
y_score = r2_score(y_test, y_pred)
```
